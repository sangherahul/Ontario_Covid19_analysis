```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Status of COVID-19 cases in Ontario.csv")
```

```python
df_clean=df["Reported Date"].str.split("T",n=1,expand=True)
df["Date"]=df_clean[0]
```

```python
fig, ax=plt.subplots(figsize=(20,10))


ax.plot(df["Date"],df["Total Cases"],marker='o',color='r')
ax.plot(df["Date"],df["Resolved"],marker='o',color='b')


ax.grid()

ticks=[]
x=1

for tick in df["Date"]:
    if x%7==0:
        ticks.append(tick)
        x=x+1
    else:
        ticks.append("")
        x=x+1
ticks

ax.set_xlabel('Time')
ax.set_ylabel('Total Cases')

plt.xticks(ticks)
plt.show()
```

### The "red" line represents the total number of Covid 19 and "blue" line represents the number of resolved cases 

Data Source: Ontario.ca
